Apply( ) Function

```python
#Classify movies based on ratings
def rating_group(rating):
    if rating >= 7.5:
        return 'Good'
    elif rating >= 6.0:
        return 'Average'
    else:
        return 'Bad'
```

```python
df['Rating_category'] = df['Rating'].apply(rating_group)
```

Matplotlib

```python
import matplotlib.pyplot as plt
```

Number of Movies per Genre

```python
genre_counts = df['Genre'].value_counts().head(10)
plt.figure(figsize=(10,5))
plt.bar(genre_counts.index, genre_counts.values, color='skyblue')
plt.xticks(rotation=45)
plt.title("Top 10 Movie Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.show()
```

Average Rating by Genre

```python
avg_rating = df.groupby('Genre')['Rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
plt.bar(avg_rating.index, avg_rating.values, color='red')
plt.xticks(rotation=45)
plt.title("Average Rating by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Rating")
plt.tight_layout()
plt.show()
```

Movies Released Per Year

```python
year_counts = df['Year'].value_counts().sort_index()
plt.figure(figsize=(12,6))
plt.bar(year_counts.index.astype(str), year_counts.values, color='green')
plt.xticks(rotation=45)
plt.title("Number of Movies Released Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.show()
```

Rating Distribution

```python
plt.figure(figsize=(8,5))
plt.hist(df['Rating'].dropna(), bins=10, color='purple', edgecolor='black')
plt.title("Distribution of Movie Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.show()
```

Votes vs Rating

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Rating'], df['Votes'], alpha=0.5, color='red')
plt.title("Votes vs Rating")
plt.xlabel("Rating")
plt.ylabel("Number of Votes")
plt.tight_layout()
plt.show()
```

 Box Plot of Ratings by Rating Category

```python
categories = df['Rating_category'].dropna().unique()
data = [df[df['Rating_category'] == cat]['Rating'].dropna() for cat in categories]

plt.figure(figsize=(10,6))
plt.boxplot(data, labels=categories)
plt.title("Boxplot of Ratings by Rating Category")
plt.xlabel("Rating Category")
plt.ylabel("Rating")
plt.tight_layout()
plt.show()
```

Top 10 Directors by Number of Movies

```python
top_directors = df['Director'].value_counts().head(10)
plt.figure(figsize=(10,5))
plt.bar(top_directors.index, top_directors.values, color='teal')
plt.xticks(rotation=45)
plt.title("Top 10 Directors by Number of Movies")
plt.xlabel("Director")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.show()
```
